## Extract WRF time series collocated with stations
 - Santa Cruz Island: LATITUDE: 33.992981 LONGITUDE: -119.716264 ELEVATION 89 m  -  WRF elevation 156.27942
 - Santa Rosa Island: LATITUDE: 33.978686 LONGITUDE: -120.078778 ELEVATION 391 m -  WRF elevation 288.4687

```python
dairt
```

```python
#  SCI
xlat =   33.992981
xlon = -119.716264
slat = dairt['XLAT'].values
slon = dairt['XLONG'].values
dist = np.sqrt((slat - xlat)**2 + (slon - xlon)**2)
ind  = np.unravel_index(np.argmin(dist, axis=None), dist.shape)
np.shape(slat),np.shape(slon),np.shape(dist),ind
sci  = dairt[:,ind[0],ind[1]]
sci  = sci.drop('XLAT')
sci  = sci.drop('XLONG')
dsci = sci.to_dataframe() - 273.15
elev = terrain[ind[0],ind[1]].values
print (elev), dsci
```

```python
#  SRI
xlat =   33.978686
xlon = -120.078778
slat = dairt['XLAT'].values
slon = dairt['XLONG'].values
dist = np.sqrt((slat - xlat)**2 + (slon - xlon)**2)
ind  = np.unravel_index(np.argmin(dist, axis=None), dist.shape)
np.shape(slat),np.shape(slon),np.shape(dist),ind
sri  = dairt[:,ind[0],ind[1]]
sri  = sri.drop('XLAT')
sri  = sri.drop('XLONG')
dsri = sri.to_dataframe() - 273.15
elev = terrain[ind[0],ind[1]].values
print (elev), dsri

```

## WRF  OBS comparison

```python
wsci  =  dsci_mon.loc['2000-01':'2018-12']
wsri  =  dsri_mon.loc['2000-01':'2018-12']

wsci
```

### Statistics

**&rarr; SCI**

```python
#   SCI
#   there are missing data

xval      = np.squeeze(osci.to_numpy())
ymod      = np.squeeze(wsci.to_numpy())
idx       = np.where(np.isfinite(xval))
xval      = xval[idx]
ymod      = ymod[idx]

res       = np.corrcoef(xval, ymod)
corr_sci  = res[0,1]
tmp       = (ymod - xval)
bias_sci  = tmp.mean()
tmp       = (ymod - xval)**2
rmse_sci  = np.sqrt(tmp.mean())
tmp       = np.abs(ymod - xval)
mabs_sci  = tmp.mean()

print ('--- Santa Cruz Island')
print ('correlation:',corr_sci,'Bias:', bias_sci, 'RMSE:', rmse_sci, 'MAE:', mabs_sci)
```

**&rarr; SRI**

```python
#   SRI
#   there are missing data

xval      = np.squeeze(osri.to_numpy())
ymod      = np.squeeze(wsri.to_numpy())
idx       = np.where(np.isfinite(xval))
xval      = xval[idx]
ymod      = ymod[idx]

res       = np.corrcoef(xval, ymod)
corr_sri  = res[0,1]
tmp       = (ymod - xval)
bias_sri  = tmp.mean()
tmp       = (ymod - xval)**2
rmse_sri  = np.sqrt(tmp.mean())
tmp       = np.abs(ymod - xval)
mabs_sri  = tmp.mean()

print ('--- Santa Rosa Island')
print ('correlation:',corr_sri,'Bias:', bias_sri, 'RMSE:', rmse_sri, 'MAE:', mabs_sri)
```
